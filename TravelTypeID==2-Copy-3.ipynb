```python
cv=StratifiedKFold(n_splits=5,shuffle=True,random_state=111)
# StratifiedKFold is a type of cross validation technique, used for highly imbalanced datasets, where it maintains the 
# actual class distribution in each fold (even when smote is applied to balance the classes). This ensures that evaluation
# will be consistent and reliable all through out.
```

```python
grid_search=GridSearchCV(estimator=model_rfc_tuned,param_grid=param_grid,cv=cv,n_jobs=-1,verbose=2,scoring='average_precision')
```
